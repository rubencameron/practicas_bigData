# ATP matches 2006

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('atp_matches_2006.csv')
df
```

```python
df['tourney_level'].value_counts()
```

```python
cuadro = df['draw_size'].value_counts()
cuadro
```

```python
df = df.dropna(how='all')
```

```python
media_ws = df['winner_seed'].mean()
media_ws
```

```python
df['winner_seed'] = df['winner_seed'].fillna(media_ws)
```

```python
df['winner_seed'].isnull().value_counts()
```
