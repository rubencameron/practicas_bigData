## Primeramente importamos las librerias a ser utilizadas para posteriormente importar el 
## archivo .csv 

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame 
```

```python
URL = 'https://raw.githubusercontent.com/florenciaortega/Prediccion-de-mora/master/DATASET_PREDICT_MORA.csv'
df = pd.read_csv(URL)
df
```

```python
df.isnull().sum() #Aquí vemos cuáles columnas contienen mas valores null
```

```python
df=df.drop(['Q_ACCESO_APP_3M', 'Q_PAGO_ELECTRONICO_6M'], axis=1) #Eliminamos las dos columnas mencionadas en el array
df                                                               #ya que ambas tienen todos sus valores NaN
```

```python
df.isnull().sum()
```

```python
nuevo_valor = df['INACTIVIDADCOMPRA'].median()
nuevo_valor
```

```python
df['INACTIVIDADCOMPRA'] = df['INACTIVIDADCOMPRA'].fillna(nuevo_valor) #Para la columna INACTIVIDADCOMPRA asignamos 
df.isnull().sum()                                                     #el valor de la media para rellenar los campos
                                                                      #faltantes
```

```python
df
```

```python
df = df.drop(['PROVINCIA'], axis=1) #Elimino la columna provincia porque no la considero necesaria
df
```

```python
df['Masculino'] = (df['SEXO'] == 'M').astype(int)
df['Femenino'] = (df['SEXO'] == 'F').astype(int)
```

```python
df = df.drop(['SEXO'], axis=1) #Creo dos columnas nuevas para colocar los valores de SEXO en modo binario
df
```

```python
df['CONTACTO_VALIDO'].value_counts()
df = df.drop(['CONTACTO_VALIDO'], axis=1)
df
```

```python
df = pd.get_dummies(df, columns=['TIENE_MAIL','USER_HOMEBANKING','DEBITOAUTOMATICO']) 
df
#Finalmente el get dummies ayuda a convertir los datos faltantes a datos binarios
#Con esto los datos quedan limpios y listos para aplicar el algoritmo de prediccion
```

# A partir de aqui empezamos con Scikit learn

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, random_state = 1234) #Esta funcion realiza un split de mi dataframe
                                                        #una parte para entreamiento y otra para testeo
```

```python
print(train.shape) #75% del dataframe para entrenamiento
print(test.shape)  #25% del dataframe para testeo
```

```python
y = df.pop('MORA_TARGET') #Es la variable dependiente que queremos averiguar
X = df
```

```python
X_train, y_train, X_test, y_test = train_test_split(X,y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head(3)
```

```python
predictors = ['ID_CLIENTE', 'EDAD', 'MESESINACTIVIDAD', 'RESUMEN_ACTIVO', 
              'CANTADICIONALES', 'LIMITE_DISP_TC', 'ANTIGUEDAD', 'INACTIVIDADCOMPRA', 
              'GRUPO_AFINIDAD', 'Masculino', 'Femenino', 'TIENE_MAIL_N', 
             'TIENE_MAIL_S', 'USER_HOMEBANKING_N', 'USER_HOMEBANKING_S', 
              'DEBITOAUTOMATICO_N', 'DEBITOAUTOMATICO_S'] 
```

```python
X_train = train[predictors].values #Mis variables independientes van a ser los valores de los predictores
print(X_train[:2])
```

```python
X_test = test[predictors].values
print(X_test[:2])
```

```python
y_train = train['MORA_TARGET'].values #Mi variable dependiente esta en la columna mencionada
print(y_train[:10])
```

```python
y_test = test['MORA_TARGET'].values
print(y_test[:10])
```

### Una vez establecidos los predictores y sus valores, ahora sigue entrenar al modelo
### primeramente usando regresion logistica

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train) 

#El metodo fit toma los datos de entrenamiento como argumentos
#The ‘fit’ method trains the algorithm on the training data, 
#after the model is initialized. That’s really all it does.
```

```python
#Luego testeamos los resultados:
```

```python
y_predict = model.predict(X_test)
y_predict[:20]
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score
metrics.accuracy_score(y_test, y_predict, normalize=True, sample_weight=None)
```

# Cross validation

```python
from sklearn.model_selection import cross_val_score
model = LogisticRegression(max_iter=1000)
#Debido a un error que tuve con la función de regresión logística, tuve que aumentar la cantidad maxima de iteraciones
#a 1000 para así evitar la ventana de mensaje que recibía.
```

```python
scores = cross_val_score(model, X_train, y_train, cv=4)
print(scores)
print(scores.mean())
```
